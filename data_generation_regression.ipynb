# Data Generation

Generation of synthetic data for regression task

## Data generation using scikit-learn's make_regression:

```python
import numpy as np
from sklearn.datasets import make_regression
```

Parameters:

```python
n_samples = 10000
n_features = 20
n_informative = 8
n_targets = 1
bias = 1
effective_rank = 20
tail_strength = 0.7
noise = 0.1
shuffle = True
coef = True
random_state = 42
```

```python
X, y, coef= make_regression(n_samples, n_features, n_informative, n_targets, bias,
                      effective_rank, tail_strength, noise, shuffle, coef,
                      random_state)
```

```python
print(coef)
```

## Gradual data collection setting

```python
batch_size = 1000
feature_limit = 5
```

```python
def feature_selection():
    
    ''' random feature selection '''
    
    return np.random.choice(n_features,size=feature_limit,replace=False)
```

```python
n_batches = int(n_samples/batch_size)
for batch in range(n_batches):
    selected_features = feature_selection()
    X_batch = data[["feature {}".format(i) for i in selected_features]][batch*batch_size:(batch+1)*batch_size].copy()
    y_batch = data["labels"][batch*batch_size:(batch+1)*batch_size].copy()
```

```python
X_batch.head()
```
